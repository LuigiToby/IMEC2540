```python
# Magnitudes y ángulos (en radianes)
mag_a, mag_b, mag_c, ang1, ang2, ang3 = symbols('mag_a mag_b mag_c ang1 ang2 ang3')

# Lista de (magnitud, ángulo)
vectores = [
    (mag_a, ang1),
    (mag_b, ang2),
    (mag_c, ang3)
]

# Cálculo de componentes, aprovechando la estructura de tuplas
componentes = [(m*cos(theta), m*sin(theta)) for m, theta in vectores]
print(componentes)
# Asignar a variables individuales
(a_x, a_y), (b_x, b_y), (c_x, c_y) = componentes

# Sustituir valores reales (convertimos grados a radianes)
subs_dict = {
    mag_a: 10,
    mag_b: 20,
    mag_c: 35,
    ang1: 90*pi/180,
    ang2: 53*pi/180,
    ang3: -32*pi/180
}

# Evaluar y simplificar
a_x_val = a_x.evalf(subs=subs_dict)
a_y_val = a_y.evalf(subs=subs_dict)
b_x_val = b_x.evalf(subs=subs_dict)
b_y_val = b_y.evalf(subs=subs_dict)
c_x_val = c_x.evalf(subs=subs_dict)
c_y_val = c_y.evalf(subs=subs_dict)

# Mostrar resultados
print(f"a_x = {a_x_val:.4f}, a_y = {a_y_val:.4f}")
print(f"b_x = {b_x_val:.4f}, b_y = {b_y_val:.4f}")
print(f"c_x = {c_x_val:.4f}, c_y = {c_y_val:.4f}")

# Suma de componentes
R_x = sum(comp[0] for comp in componentes)
R_y = sum(comp[1] for comp in componentes)

R_mag = (R_x**2 + R_y**2)**0.5
R_ang = atan2(R_y, R_x)  # en radianes

R_x_val = R_x.evalf(subs=subs_dict)
R_y_val = R_y.evalf(subs=subs_dict)
R_mag_val = R_mag.evalf(subs=subs_dict)
R_ang_val = R_ang.evalf(subs=subs_dict)

print(f"R_x = {R_x_val:.4f}, R_y = {R_y_val:.4f}")
print(f"|R| = {R_mag_val:.4f}")
print(f"Ángulo = {(R_ang_val*180/pi):.2f}°") # Ángulo de la resultante con cuidado de pasarlo a grados

# Proyecciones sobre R
proy_a = ((a_x*R_x + a_y*R_y) / R_mag).evalf(subs=subs_dict)
proy_b = ((b_x*R_x + b_y*R_y) / R_mag).evalf(subs=subs_dict)
proy_c = ((c_x*R_x + c_y*R_y) / R_mag).evalf(subs=subs_dict)

# Mostrar
print(f"Proyección de a sobre R: {proy_a:.4f}")
print(f"Proyección de b sobre R: {proy_b:.4f}")
print(f"Proyección de c sobre R: {proy_c:.4f}")
```

```python
# Magnitudes y ángulos (en radianes)
v1, v2, ang1, ang2 = symbols('v1 v2 ang1 ang2')
# Lista de (magnitud, ángulo)
vectores = [
    (v1, ang1),
    (v2, ang2)
]

# Cálculo de componentes, aprovechando la estructura de tuplas
componentes = [(m*cos(theta), m*sin(theta)) for m, theta in vectores]
print(componentes)
# Asignar a variables individuales
(v_1x, v_1y), (v_2x, v_2y) = componentes

# Sustituir valores reales (convertimos grados a radianes)
subs_dict = {
    ang2: atan2(4, 3),
    ang1: 30*pi/180,
}

# Evaluar y simplificar
v_1x_val = v_1x.subs(subs_dict)
v_1y_val = v_1y.subs(subs_dict)
v_2x_val = v_2x.subs(subs_dict)
v_2y_val = v_2y.subs(subs_dict)

# Mostrar resultados
print(f"v_1x = {v_1x_val}, v_1y = {v_1y_val}")
print(f"v_2x = {v_2x_val}, v_2y = {v_2y_val}")

# --- Operaciones vectoriales ---

# Suma
sum_x = simplify(v_1x_val + v_2x_val)
sum_y = simplify(v_1y_val + v_2y_val)

# Resta
diff_x = simplify(v_1x_val - v_2x_val)
diff_y = simplify(v_1y_val - v_2y_val)

# Producto cruz (en 2D → escalar z)
cross_val = simplify(v_1x_val*v_2y_val - v_1y_val*v_2x_val)

# Producto punto
dot_val = simplify(v_1x_val*v_2x_val + v_1y_val*v_2y_val)

# Resultados
print("v1 + v2 =", (sum_x, sum_y))
print("v1 - v2 =", (diff_x, diff_y))
print("v1 x v2 =", cross_val)  # escalar
print("v1 · v2 =", dot_val)
```

```python
t = symbols('t')

# Define dynamic symbols
l = dynamicsymbols('l')
theta = dynamicsymbols('theta')

# Unit vectors as dynamicsymbols (symbols only)
e_r = dynamicsymbols('e_r')
e_theta = dynamicsymbols('e_theta')

# Position vector
r = l * e_r

# Angular velocity
theta_dot = theta.diff(t)

# Derivative substitution rules for unit vectors and variables
rules = {
    Derivative(e_r, t): theta_dot * e_theta,
    Derivative(e_theta, t): -theta_dot * e_r,
    Derivative(l, t, 2): 0,
    Derivative(theta, t, 2): 0
}

# Velocity and acceleration symbolic expressions
v = r.diff(t).subs(rules).simplify()
a = v.diff(t).subs(rules).simplify()

# Function to compute magnitude from symbolic vector components
def magnitude(vec, basis_vectors):
    mag_sq = 0
    for b in basis_vectors:
        coeff = vec.coeff(b)
        mag_sq += coeff**2
    return sqrt(mag_sq).simplify()

basis = [e_r, e_theta]

# Substitute numerical values (consistent units, e.g. meters)
subs_numeric = {
    Derivative(e_r, t): (60 * pi / 180) * e_theta,
    Derivative(e_theta, t): -(60 * pi / 180) * e_r,
    Derivative(l, t): -0.15,
    Derivative(theta, t): 60 * pi / 180,
    l: 0.375,  # 375 mm = 0.375 m
}

v_num = v.subs(subs_numeric).simplify()
a_num = a.subs(subs_numeric).simplify()

v_mag = magnitude(v_num, basis)
a_mag = magnitude(a_num, basis)

print("Velocity vector:", v_num)
print("Acceleration vector:", a_num)
print("Velocity magnitude:", v_mag.evalf())
print("Acceleration magnitude:", a_mag.evalf())
```
