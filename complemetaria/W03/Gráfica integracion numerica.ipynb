```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función
def funcion(x):
    return x**3-5*x**2+2*x+8

# Crear un rango de valores de x
x = np.linspace(0, 5, 1000)

# Calcular los valores de y correspondientes
y = funcion(x)


# Graficar la función
plt.plot(x, y, label='f(x)', color='blue')

# Rellenar el área bajo la curva
plt.fill_between(x, y, where=(y > 0), color='skyblue', alpha=0.4)
plt.fill_between(x, y, where=(y < 0), color='skyblue', alpha=0.4)
# Añadir etiquetas y título
plt.xlabel('x')
plt.ylabel('y')
plt.title('Área bajo la curva de f(x)')
plt.grid(True)
# Mostrar leyenda
plt.legend()

# Mostrar la gráfica
plt.show()
```
