```python
# Escriba su función para la solucion del problema de N masas con N distancias

from sympy import symbols

def centro_de_masa_sympy(masas, posiciones):
    if len(masas) != len(posiciones):
        raise ValueError("Listas deben tener la misma longitud")
    









    M = sum(masas)
    
    
    
    
    
    
    
    
    
    
    
    if M == 0:
        raise ValueError("La suma de masas no puede ser 0")
    
    
    
    
    
    return sum(m*x for m, x in zip(masas, posiciones)) / M

# Ejemplo:








m1, m2 = symbols('m1 m2')




x1, x2 = symbols('x1 x2')










print(centro_de_masa_sympy([m1, m2], [x1, x2]).subs({m1:5, x1:0, m2:7, x2:2}))
```

```python
# Resuelva el problema aquí


import sympy as sp

# Variables
x, L, M = sp.symbols('x L M', positive=True, real=True)










# densidad lineal
lambda_ = M / L








# Numerador y denominador de la fórmula




num = sp.integrate(x * lambda_, (x, 0, L))
den = sp.integrate(lambda_, (x, 0, L))











x_cm = sp.simplify(num / den)
print("x_CM simbólico:", x_cm)

# Sustitución de valores L=3 m, M=4 kg
x_cm_val = x_cm.subs({L: 3, M: 4})
print("x_CM numérico:", float(x_cm_val))
```

```python
# Haga la función para el caso de la integración del solido de revolución



import sympy as sp

def cm_solido_revolucion(fx, a, b, axis='x'):
    x = sp.symbols('x', real=True)
    fx = sp.sympify(fx)



    if axis == 'x':
        V  = sp.integrate(sp.pi*fx**2, (x, a, b))
        xcm = sp.integrate(x*sp.pi*fx**2, (x, a, b)) / V
        return (sp.simplify(xcm), sp.Integer(0), sp.Integer(0))





    elif axis == 'y':
        V  = sp.integrate(2*sp.pi*x*fx, (x, a, b))
        ycm = sp.integrate(sp.pi*x*fx**2, (x, a, b)) / V
        return (sp.Integer(0), sp.simplify(ycm), sp.Integer(0))






    else:
        raise ValueError("axis debe ser 'x' o 'y'")





x = sp.symbols('x', real=True)





cm1 = cm_solido_revolucion(1 - x**2, 0, 1, axis='x')

cm2 = cm_solido_revolucion(1 - x**2, 0, 1, axis='y')

print("CM eje x:", cm1)
print("CM eje y:", cm2)
```

## Scotch-yoke

```python
import math
import sympy as sp
from sympy.physics.mechanics import ReferenceFrame, Point, dynamicsymbols

# ---- símbolos y frames ----
t = sp.symbols('t')
theta = dynamicsymbols('theta')
omega, alpha, r, d = sp.symbols('omega alpha r d', real=True)

N = ReferenceFrame('N')
B = ReferenceFrame('B')
B.orient_axis(N, theta, N.z)  # B gira con θ






O = Point('O'); O.set_vel(N, 0)
A = O.locatenew('A', r*B.x)




# ---- cinemática en marco rotante ----
A.v2pt_theory(O, N, B)  # velocidad
A.a2pt_theory(O, N, B)  # aceleración






# Proyección en x e imposición de ω, α instantáneos
vAx = sp.simplify(A.vel(N).dot(N.x)).subs({theta.diff(t): omega})
aAx = sp.simplify(A.acc(N).dot(N.x)).subs({theta.diff(t): omega,
                                           theta.diff(t,2): alpha})

print("v_Ax simbólico =", vAx)
print("a_Ax simbólico =", aAx)

# ---- valores numéricos ----
theta_deg = 65
omega_val = 5.0           # rad/s
alpha_val = 8.0           # rad/s^2
r_mm = 10.0               # mm
d_mm = 45.0               # mm (no afecta v/a)
theta_val = math.radians(theta_deg)
r_m = r_mm / 1000.0

num_subs = {
    theta: theta_val,
    omega: omega_val,
    alpha: alpha_val,
    r: r_m,
    d: d_mm/1000.0
}

vPx = float(vAx.subs(num_subs))
aPx = float(aAx.subs(num_subs))

# Unidades convenientes
vPx_mm_s  = vPx * 1000.0
aPx_mm_s2 = aPx * 1000.0

print("v_Px [m/s], a_Px [m/s^2]  =", vPx, aPx)
print("v_Px [mm/s], a_Px [mm/s^2] =", vPx_mm_s, aPx_mm_s2)
```

## Slider-crank

```python
# Re-run after state reset

import sympy as sp

t = sp.symbols('t')
r, l, w0 = sp.symbols('r l w0', positive=True, real=True)
theta = sp.symbols('theta', real=True)







xB = r*sp.cos(theta)
yB = r*sp.sin(theta)
xA = sp.simplify(xB - sp.sqrt(l**2 - yB**2))








dx = sp.simplify(xB - xA)
dy = sp.simplify(yB)



vBx = -r*w0*sp.sin(theta)
vBy =  r*w0*sp.cos(theta)





xAdot, wAB = sp.symbols('xAdot wAB', real=True)
kx_rBA_x = -dy
kx_rBA_y =  dx





eq_vx = sp.Eq(vBx - xAdot, wAB * kx_rBA_x)
eq_vy = sp.Eq(vBy,         wAB * kx_rBA_y)




sol_vel = sp.solve([eq_vx, eq_vy], (xAdot, wAB), simplify=True, dict=True)[0]
xAdot_expr = sp.simplify(sol_vel[xAdot])
wAB_expr = sp.simplify(sol_vel[wAB])




aBx = -r*w0**2*sp.cos(theta)
aBy = -r*w0**2*sp.sin(theta)




xAddot, aAB = sp.symbols('xAddot aAB', real=True)




eq_ax = sp.Eq(aBx - xAddot, aAB * kx_rBA_x - wAB_expr**2 * dx)
eq_ay = sp.Eq(aBy,          aAB * kx_rBA_y - wAB_expr**2 * dy)






sol_acc = sp.solve([eq_ax, eq_ay], (xAddot, aAB), simplify=True, dict=True)[0]
xAddot_expr = sp.simplify(sol_acc[xAddot])
aAB_expr = sp.simplify(sol_acc[aAB])




wAB_expr_s = sp.simplify(sp.trigsimp(wAB_expr))
aAB_expr_s = sp.simplify(sp.trigsimp(aAB_expr))





wAB_expr_s, aAB_expr_s
```
