```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, pi, sqrt

# Parámetros simbólicos
l, h_c = symbols('l h_c', real=True, positive=True)

# Definimos parámetros fijos de ejemplo
l_val = 73   # por ejemplo, diámetro de la varilla 1 cm
h_c_val = 100  # altura constante de la pieza (ejemplo)

# Definimos función de r(t) y v(r)
def r_t(t, l_val):
    return 2*np.sqrt((l_val/2)**2 + (t/2)**2)

def v_r(r, h_c_val):
    return np.pi * r**2 * h_c_val

# Intervalos en pasos de 1.25 m
t_values = np.arange(0, 100, 1.25)  # por ejemplo hasta 10 metros
r_values = r_t(t_values, l_val)
v_values = v_r(r_values, h_c_val)

# Graficar
plt.figure(figsize=(7,5))
plt.plot(r_values, v_values, marker='o', label="v vs r (discreto)")
plt.xlabel("Radio r [m]")
plt.ylabel("Volumen v [m³]")
plt.title("Gráfico de v vs r en intervalos de 1.25 m de avance")
plt.grid(True)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
from sympy import symbols, pi, sqrt

# Variables simbólicas
t, l, h_c = symbols('t l h_c', real=True, positive=True)

# Definiciones simbólicas
r_t = 2*sqrt((l/2)**2 + (t/2)**2)
v_r = pi * r_t**2 * h_c

# Parámetros fijos en mm
l_val = 73     # diámetro de la varilla [mm]
h_c_val = 100  # altura de la pieza [mm]

# Intervalos de avance (ejemplo: 1.25 mm por paso)
t_step = 1.25   # paso del tornillo
t_max = 200     # límite para graficar
t_values = [i*t_step for i in range(int(t_max/t_step)+1)]

# Evaluar r y v
r_values = [float(r_t.subs({t: tv, l: l_val})) for tv in t_values]
v_values = [float(v_r.subs({t: tv, l: l_val, h_c: h_c_val})) for tv in t_values]

# Graficar
plt.figure(figsize=(7,5))
plt.plot(r_values, v_values, marker='o', label="v vs r (discreto)")
plt.xlabel("Radio r [mm]")
plt.ylabel("Volumen v [mm³]")
plt.title("Gráfico de v vs r en intervalos del tornillo")
plt.grid(True)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
from sympy import symbols, sqrt, pi, lambdify
import numpy as np

# Parámetros
l_val = 73     # mm
h_c_val = 150  # mm
t_max = 27     # mm
paso = 1.25    # mm por vuelta

# Definiciones simbólicas
t, l, h_c = symbols('t l h_c', real=True, positive=True)
r = (sqrt((l/2)**2 + (t/2)**2))*2          # radio
v = pi * r**2 * h_c                # volumen

# Funciones numéricas
r_func = lambdify((t, l), r, 'numpy')
v_func = lambdify((t, l, h_c), v, 'numpy')

# Valores discretos de t
t_values = np.arange(0, t_max + paso, paso)
vueltas = t_values / paso
v_values = v_func(t_values, l_val, h_c_val)

# Gráfico
plt.figure(figsize=(7,5))
plt.plot(vueltas, v_values, marker='o', label="Volumen vs vueltas")
plt.xlabel("Vueltas del tornillo")
plt.ylabel("Volumen [mm³]")
plt.title("Volumen del cilindro vs número de vueltas")
plt.grid(True)
plt.legend()
plt.show()
```
